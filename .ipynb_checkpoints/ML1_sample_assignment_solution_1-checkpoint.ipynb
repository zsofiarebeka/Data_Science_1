```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(20240221)

# Settings
n = 500
R = 5000

# Define true function expecting the feature matrix as input
def f(X):
    return X[:, 0]**3 - 3.5*X[:, 0]**2 + 3*X[:, 0]

# Specific values to evaluate
x_values = np.array([[0.1, 0, 0], [1.1, 0, 0]])

# Initialize
results = []

# Monte Carlo simulation
for _ in range(R):
    
    # Generate data
    X1, X2, X3 = [np.random.uniform(0, 2, n) for _ in range(3)]
    X = np.column_stack((X1, X2, X3))
    epsilon = np.random.normal(0, 1, n)
    Y = f(X) + epsilon

    # Model 1
    model1 = LinearRegression().fit(X, Y)
    pred1 = model1.predict(x_values)

    # Model 2
    poly2 = PolynomialFeatures(3)
    X_poly2 = poly2.fit_transform(X[:, 0].reshape(-1, 1))
    model2 = LinearRegression().fit(X_poly2, Y)
    pred2 = model2.predict(poly2.transform(x_values[:, 0].reshape(-1, 1)))

    # Model 3
    poly3 = PolynomialFeatures(3)
    X_poly3 = poly3.fit_transform(X)
    model3 = LinearRegression().fit(X_poly3, Y)
    pred3 = model3.predict(poly3.transform(x_values))

    # Store results
    results.append([pred1, pred2, pred3])

# Compute bias, variance and MSE
results = np.array(results)
bias = np.mean(results, axis=0) - f(x_values)
variance = np.var(results, axis=0)
mse = np.mean((results - f(x_values))**2, axis=0)

# Organize results into a table
df = pd.DataFrame({
    'x1': np.tile(x_values[:, 0], 3),
    'model': np.repeat(['Model 1', 'Model 2', 'Model 3'], 2),
    'bias': bias.flatten(),
    'variance': variance.flatten(),
    'MSE': mse.flatten()
})
df.sort_values(by=["x1", "model"])

# This is natural that one model performs well on one domain but doesn't perform well on another domain --> higher RMSE
# Here, we don't have the epsilon, the irreducible part of the error
# Diff between MSE and MSPE is the variance
```

```python
np.random.seed(20240221)

# Settings
n = 500
R = 10

# Specific values to evaluate
first_dim = np.arange(0, 2.05, 0.05)
x_values_to_plot = np.column_stack((first_dim, np.zeros(len(first_dim)), np.zeros(len(first_dim))))

# Initialize
results_to_plot = []

# Monte Carlo simulation
for _ in range(R):
    
    # Generate data
    X1, X2, X3 = [np.random.uniform(0, 2, n) for _ in range(3)]
    X = np.column_stack((X1, X2, X3))
    epsilon = np.random.normal(0, 1, n)
    Y = f(X) + epsilon


    # Model 1
    model1 = LinearRegression().fit(X, Y)
    pred1 = model1.predict(x_values_to_plot)


    # Model 2
    poly2 = PolynomialFeatures(3)
    X_poly2 = poly2.fit_transform(X[:, 0].reshape(-1, 1))
    model2 = LinearRegression().fit(X_poly2, Y)
    pred2 = model2.predict(poly2.transform(x_values_to_plot[:, 0].reshape(-1, 1)))

    # Model 3
    poly3 = PolynomialFeatures(3)
    X_poly3 = poly3.fit_transform(X)
    model3 = LinearRegression().fit(X_poly3, Y)
    pred3 = model3.predict(poly3.transform(x_values_to_plot))

    # Store results_to_plot
    results_to_plot.append([pred1, pred2, pred3])

# Plotting predictions
import matplotlib.pyplot as plt

# Create a figure and axis objects
fig, axs = plt.subplots(1, 3, sharey=True)

# Iterate over each subchart
for i, ax in enumerate(axs):
    # Plot each line for the current subchart
    for j in range(10):
        ax.plot(x_values_to_plot[:, 0], results_to_plot[j][i], color="navy", alpha=0.5)

    for x_val in x_values[:, 0]:
        ax.axvline(x_val, color='gray', linestyle='dashed', label=f'x = {x_val}')

    ax.plot(x_values_to_plot[:, 0], f(x_values_to_plot), color="darkred", linewidth=2)
    
    ax.set_title(f'Model {i+1}')

plt.show()


# True model is not always the best model but it's always unbiased
```

```python
# Generating a new sample

import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

np.random.seed(20240228)

# Settings
n = 1000

# Define true function expecting the feature matrix as input
def f(X):
    return X[:, 0]**3 - 3.5*X[:, 0]**2 + 3*X[:, 0]

# Initialize
results = []

# Initialize
results_to_plot = []

# Generate data
X1, X2, X3 = [np.random.uniform(0, 2, n) for _ in range(3)]
X = np.column_stack((X1, X2, X3))
epsilon = np.random.normal(0, 1, n)
y = f(X) + epsilon

# TRain-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5)

# Estimate model 1
# We have to reshape it into 2D
# Reshape works for any kind of arrays
# Number of rows (-1 refers to the OG number of rows), number of columns(1 refers to 1 column)
# Reshape just wants to have one column
# the fit methods just estimates parameters from the data
# the fit method used to compute parameters
# LinReg has slope and intercept parameter which is learned during this fit process
model1 = LinearRegression().fit(X_train[:, 0].reshape(-1, 1), y_train)
pred1 = model1.predict(X_test[:, 0].reshape(-1, 1))

# Model 2
# poly2 is a 3-degree polynomial
# Change this poly2 object to actually 1D
# Transform the other method of this preprocessing step: fit_transform: it fits and transforms
# We only transform the polynomials, but only X1
poly2 = PolynomialFeatures(3)
X_poly2 = poly2.fit_transform(X_train[:, 0].reshape(-1,1))
model2 = LinearRegression().fit(X_poly2, y_train)
pred2 = model2.predict(poly2.transform(X_test[:, 0].reshape(-1,1)))

# Model 3
# We transform all polynomials
poly3 = PolynomialFeatures(3)
X_poly3 = poly3.fit_transform(X_train)
model3 = LinearRegression().fit(X_poly3, y_train)
pred3 = model3.predict(poly3.transform(X_test))

# Compute MSPE
mspe1 = np.mean((pred1 - y_test)**2)
mspe2 = np.mean((pred2 - y_test)**2)
mspe3 = np.mean((pred3 - y_test)**2)

#np.mean(y_test) is around 0.4, the MSPE is quite large
pd.DataFrame({
    "model": ["Model 1", "Model 2", "Model 3"],
    "MSPE": [mspe1, mspe2, mspe3]
})
# the MSPE is around 1
# The variance for this exercise is actually 1
# MSPE contains the variance
```

```python
# Lasso simulation
import numpy as np
from sklearn.linear_model import Lasso

def f(X): 
    # f(X) = x1 + x2
    # : minden sor a nullás oszlopból és minden sor az egyes oszlopból
    return X[:, 0] + X[:, 1]

n = 20
R = 1000
# [[]] refers to
# We are evaluating at 0, 0
x_to_evaluate = np.array([[0, 0]])
alphas_to_try = np.arange(0.01, 0.5, 0.02)

# Monte Carlo
# empty array contains R rows and 1 column
results = np.empty((R, len(alphas_to_try)))
for _ in range(R):
    
    # Generate data
    X1, X2 = [np.random.uniform(0, 1, n) for _ in range(2)]
    X = np.column_stack((X1, X2))
    # variance is going to be 2
    epsilon = np.random.normal(0, 2, n)
    Y = f(X) + epsilon


    # Estimate LASSO
    for id_a, a in enumerate(alphas_to_try):
        lasso = Lasso(alpha = a).fit(X, Y)
        pred = lasso.predict(x_to_evaluate)
        results [_, id_a] = pred

results
# Compute bias, variance and mse
bias = np.mean(results - f(x_to_evaluate), axis = 0) 
variance = np.var(results, axis = 0)
mse = np.mean(((results - f(x_to_evaluate))**2), axis = 0)
bias, variance, mse
```

```python
import matplotlib.pyplot as plt
plt.plot(alphas_to_try, bias**2, label = "Bias^2")
plt.plot(alphas_to_try, variance, label = "Variance")
plt.plot(alphas_to_try, mse, label = "MSE")
plt.xlabel('Penalty parameter')
plt.legend()
plt.show
```
